```python
import pandas as pd
```

```python
#read dataframe
df = pd.read_csv('order_brush_order.csv')
df.info()
```

```python
df.head()
```

```python
#ubah event time jadi index buat pakai pd.Grouper 
df = df.set_index(pd.DatetimeIndex(df['event_time']))
```

```python
df = df.drop(columns = 'event_time')
```

```python
df
```

```python
df_group = df.groupby(['shopid','userid',pd.Grouper(freq='60min')]).count()
```

```python
df_group
```

```python
df_group = df_group.rename(columns = {'orderid':'n_order'})
```

```python
df_group
```

```python
#cari n_order yang lebih dari sama dengan 3
df_suspect = df_group[df_group['n_order'] > 2]
```

```python
df_suspect
```

```python
df_suspect = df_suspect.reset_index()
```

```python
df_suspect
```

```python
print('Jumlah toko yang dicurigai {}'.format(len(df_suspect['shopid'].unique())))
print('Jumlah user yang order lebih dari 3 dalam 1 jam {}'.format(len(df_suspect['userid'].unique())))
```

```python
#buat list untuk shop id yang memiliki user id > 1 dan n order > 2
list1 = []
df_suspect.groupby('shopid')['userid'].apply(lambda x: list1.append(x.values))
len(df_suspect['shopid'].unique()) == len(list1)
```

```python
list1
```

```python
#tambah '&' di tengah"
list2 = []
for i in list1:
    if len(i)>1:
        list2.append('&'.join(str (j) for j in i))
    else:
        list2.append(''.join(str (i) for i in i))
        
list2
```

```python
#buat userid untuk yang brushing
userid_brush = pd.DataFrame({'userid' : list2})
userid_brush
```

```python
#buat shopid untuk yang brushing
shopid_brush = pd.DataFrame({'shopid' : df_suspect['shopid'].unique()})
shopid_brush
```

```python
#pengecekan jumlah row data sama
len(shopid_brush) == len(userid_brush)
```

```python
#append 2 datafram
order_brushing = pd.DataFrame({'shopid' : df_suspect['shopid'].unique(), 'userid' : list2})
```

```python
order_brushing
```

```python
df_all = pd.DataFrame({'shopid': df['shopid'].unique(), 'userid':0})
```

```python
df_all
```

```python
#shop id yang tidak melakukan order brushing
not_order_brushing = df_all[~df_all['shopid'].isin(order_brushing['shopid'])]
not_order_brushing
```

```python
#gabungkan order brushing dengan yang tidak
result = pd.concat([df_all[~df_all['shopid'].isin(order_brushing['shopid'])], order_brushing])
```

```python
result
```

```python
#ubah ke bentuk csv
csv = result.to_csv('hasil_akhir.csv')
```
