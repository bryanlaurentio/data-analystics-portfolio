```python
#define public holidays
holidays = ['2020-03-08', '2020-03-25', '2020-03-30', '2020-03-31']
weekmask = [1, 1, 1, 1, 1, 1, 0]
```

```python
#ambil date pada pick, 1st attempt
pick = []
for i in df['pick']:
    pick.append(i.date())
    
first = []
for i in df['1st_deliver_attempt']:
    first.append(i.date())
```

```python
#ambil date pada second
second = []
for i in df['2nd_deliver_attempt']:
    if pd.isnull(i):
        second.append(np.datetime64('1970-01-01'))
    else:
        second.append(i.date())
```

```python
df['first_attempt'] = np.busday_count(pick, first, weekmask, holidays)
```

```python
df.head()
```

```python
df['second_attempt'] = np.busday_count(first, second, weekmask, holidays)
```

```python
df.head()
```

```python
df['is_late'] = 0
```

```python
#change data type of sla 
df['sla'] = df['sla'].values.astype(int)
```

```python
#2nd Attempt must be no later than 3 working days after the 1st Attempt
df['is_late'] = (df['second_attempt'] > 3) | (df['first_attempt']> df['sla'])
```

```python
df.head(20)
```

```python
#mat diskrit, false = 0, true = 1
df['is_late'] = df['is_late'].values.astype(int)
```

```python
df.head(20)
```

```python
result = pd.DataFrame({'orderid':df['orderid'],
                      'is_late':df['is_late']})
```
